## 1. Crear un data view

Para poder trabajar en Kibana con los datos almacenados en los índices de Elasticserach lo primero que tenemos que hacer es crear un data view.

Un data view es un objeto de Kibana que nos permite descubrir, explorar, consultar y analizar los documentos indexados en uno o mas índices en eleasticsearch. 

### 1a. Accedemos a la sección de gestión de Data Views de Kibana

Para acceder a la sección de Data Views primero  tenemos que entrar en la sección de Stack Management:

Menu > Management > Stack Management

<img src="../images/els/stackmanagement.png" alt="ingest process"/>

Stack Management es la sección para gestionar todos los objetos de Elastic Stack como índices, clusters, licencias, data views, espacios, licencias, etc. 
Por ahora vamos a centrárnos en la gestión de Data Views.

Para crear nuestro data view accedemos a las sección de gestión de data views: 

Kibana > Data Views

<img src="../images/els/kibanadataview.png" alt="ingest process"/>

En esta página encontramos el listado de los Data Views ya creados. 

### 2a. Creamos el Data View para el dataset de trips que acabamos de ingestar.

Para crear nuestro primer data view clickamos en el botón Create Data View.

Un Data View se puede crear para el conjunto de documentos de uno o varios índices, por lo que es necesario indicar el índice o índices que van a ser la fuente de datos de esta data view. Para ello podemos especificar o bien el nombre del índice sobre el que crear el Data View o en el caso de querer crear el Data View sobre varios índices el patrón o expresión regular que tiene que cumplirse sobre el nombre de los índices.

En nuestro caso vamos a utilizar el índice que hemos creado antes en la ingesta de datos, por lo que en el campo name introducimos el nombre del índice: trips.

Puesto que el data set que hemos ingestado representa una serie de eventos en el tiempo, es por tanto una serie temporal, podemos indicar cual es el campo de tipo fecha que almacena esta información temporal introduciendo en el campo Timestamp field el campo StartDate. 

Una vez informado correctamente el formulario clicamos en Create Data View.

La pagina que nos muestra a continuación contiene la información del Data View que hemos creado, donde podemos ver todos los campos de nuestro data view y su información:

* Nombre del campo.
* Tipo del dato que almacena el campo.
* Si tiene algún formato definido.
* Si se puede buscar.
* Si se puede utilizar en una agregación.
* Si es un campo excluido o no.

A parte de los campos de los documentos, podemos ver que también añade como un campo más los metadatos del índice, como el _index, _score, _source, _type, etc. 

<img src="../images/els/tripsdataview.png" alt="ingest process"/>

### 3.2. Filtro de tiempo

Vamos a profundizar en el uso del filtro de tiempo que utilizamos perviamente para mostrar de forma correcta los 
datos.

Este filtro siempre acturará sobe el campo fecha que seleccionamos al crear el data view que estamos consultando.

<img src="../images/els/filtroFechaDataview.png" alt="ingest process"/>

Tenemos varias opciones para configurar este filtro:
* **Filtro rápido**: haciendo clic en el icono superior de la derecha con forma de calendario nos aparecerá el filtro rápido donde podermos seleccionar un rango de tiempo desde o a partir de la fecha actual para el que queremos mostrar datos. Además nos da las opciones más comunes y las últimas opciones usadas.
* **Filtro relativo**: Similar al filtro rápido.
* **Filtro absoluto**: Podemos indicar una fecha concreta de inicio desde la cuál mostrar los datos.
* **Now**: filtra los datos a partir de la fecha actual. Es muy util para índices que representan streams de datos y sólo queremos consultar los nuevos eventos.


### 3.2. Barra de búsqueda

La barra de búsqueda nos permitirá introducir una consulta en lenguaje KQL (Kibana Query Language) para filtrar los datos.

KQL es un lenguaje muy sencillo que nos permite realizar los siguientes tipos de consultas:
* **Terms quey**: Los términos están separados por un espacio y sólo se requiere que un término cumpla la sentencia. Ejemmplo: nombre_campo:term1 term2 term3 
    * Si queremos que cumpla una frase entera la encerraremos entre comillas. Ejemplo: nombre_campo:"frase a buscar".
* **Boolean query**: Podemos contatenar más de una condición de búsqueda con las palabras reservadas and y or:
    * nombre_campo_a:termino1 and nombre_campo_b:termino2
    * nombre_campo_a:termino1 or nombre_campo_b:termino2
    * nombre_campo_a:termino1 and (nombre_campo_b:termino2 or nombre_campob:termino3)
    * nombre_campo_a:(termino1 or termino2)
    * not nombre_campo_a:termino1
* **Range query**: Podemos filtar por rango de valores en campos de tipo numérico y fecha. Los operadores permitdos son los siguientes: >, >=, <, y <=
    * nombre_campo_numerico > 100
    * nombre_campo_timestamp > "2021-01-02T21:55:59"
    * nombre_campo_timestamp >= "2021-01"
    * nombre_campo_timestamp < now-1d (para ver las expresiones matemáticas soportadas consultar este enlace: https://www.elastic.co/guide/en/elasticsearch/reference/8.4/common-options.html#date-math).
* **Exist query**: para saber si un campo existe:
    * nombre_campo:*
* **Wildcard query**: 
    * valores del campo: nombre_campo:win*
    * nombe del campo: nombre.campo.*:windows10
* **Nested field queyr**: Para consultar campos anidados, tenemos diferentes opciones en función del resultado que queramos obtener:
    * Resultado de la búsqueda en un único documento: nombre_campo:{nombre_campo_anidado1:termino1 and nombre_campo_anidado2:termino2} devolvera los documentos que cumplan a la vez la condición de búsqueda en el mismo documento anidado.
    * Resultado de la búsqueda en diferentes documentos: nombre_campo:{nombre_campo_anidado1:termino1} and nombre_campo:{nombre_campo_anidado2:termino2}

### 3.3. Filtros de campos

Podemos añadir tantos filtros de campos como queramos. Hay que tener en cuenta que los filtros se concatenand con el operador and. 

<img src="../images/els/filtroCampoDataview.png" alt="ingest process"/>

Al crea el filtro seleccionamos el campo por el que quermos filtrar, la condición a cumplir y los valores por los que se quiere filtrar. Además podemos ponerles un nombre que nos permitirá encontrarlos con mas facilidad.

Podemos optar por realizar una consulta con el DSL de Elasticsearch, para ello clicaremos en el enlace "Edit as Query DSL".

Una vez creados, se pueden deshabilitar y volver a habilitar. Para ello sólo basta con clicar sobre el filtro y seleccionar la opción deseada.

También permite obterner el resultado inverso del filtro, es decir los documentos que no cumplen el filtro. Igual que antes sólo es necesario clicar sobre el filtro y seleccionar la opción "Exclude results".


### 3.4. Crear nuevos campos

Otra operación muy útil a la hora de descubrir y preparar los datos es crear nuevos campos en función de los valores de los campos del documento, por ejemplo, crea un nuevo campo que se llame duración y se calcule restando el valor del campo hora_llegada del valor del campo hora_salida.

Para realizar este tipo de operaciones para crear el nuvo valor utilizaremos el lenguaje de scripting painless de Elasticsearch que ya hemos visto con antelación.

Para crear el nuveo campo clicamos en el botón que se encuentra a la izaquierda en la parte inferior de la página que pone "Add a field". 

Se nos abrirá un menú de creación.

<img src="../images/els/addFieldDataView.png" alt="ingest process"/>

Tenemos que elegir el nombre del campo y el type.

Habilitaremos la opción set value e undicaremos en la caja que se habilita el script que genera el nuevo valor. De forma dinámica en la parte derecha se genera una preview sobre un documento existente. En esta preview podemos ver si el script que estamos generando devuelve el valor esperado o si por el contrario ocurre algún error de codificación.


### 3.4. Guardar consultas

Podemos guardar nuestras consúltas para utilizarlas posteriormente para gnerar reportes, usarlas para crear visualizaciones o dashboards. 
Cuando se guarda una consulta, se almacena el texto de la query, los filtros, y la vista actual en el apartado Discover, incluyendo las columnas seleccionadas, el sentido de la ordenación y el data view.

Para guardar una consulta sólo hay qeu clicar en el boton "Save" de la barra de herramientas superior y darle un nombre.
